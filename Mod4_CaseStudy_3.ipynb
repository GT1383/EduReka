##### 1. You are given a dataset, which is present in the LMS, containing the number of hurricanes occurring in the United States along the coast of the Atlantic. Load the data from the dataset into your program and plot a Bar Graph of the data, takingthe Year as the x-axis and the number of hurricanes occurring as the Y-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
data = pd.read_csv("Hurricanes.csv")

# Step 2: Plot a bar graph
plt.figure(figsize=(10,6))
plt.bar(data["Year"], data["Hurricanes"], color="skyblue", edgecolor="black")

# Step 3: Add labels and title
plt.xlabel("Year")
plt.ylabel("Number of Hurricanes")
plt.title("Number of Hurricanes along the Atlantic Coast")

# Step 4: Show the plot

plt.show()
```

##### 2. The dataset given, records data of city temperatures over the years 2014 and 2015. Plot the histogram of the temperatures over this period for the cities of San Francisco and Moscow.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
data = pd.read_csv("CityTemps.csv")

# Step 2: Plot histograms for both cities

plt.hist(data["San Francisco"], bins=20, alpha=0.7, label="San Francisco", color="skyblue", edgecolor="black")
plt.hist(data["Moscow"], bins=20, alpha=0.7, label="Moscow", color="pink", edgecolor="black")

# Step 3: Add labels and title
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.title("Temperature Distribution (2014â€“2015) for San Francisco and Moscow")
plt.legend()

# Step 4: Show the plot
plt.show()
```

```python
# 5.1: Draw a Simple plot
# 5.2: Configure the line and markers in a simple plot
# 5.3: configure the axes
# 5.4: Give the title of Graph & labels of the x axis and y axis

import matplotlib.pyplot as plt

# Data points
X = [1, 2, 3, 4]
y = [20, 21, 20.5, 20.8]

plt.plot(
    X, y,
    color="blue",      
    linestyle="--",      
    marker="o",         
)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis (Values)")
plt.title("Configured Line and Marker Plot")

# Show the plot
plt.show()
```

```python
# 5.5: Give error bar if y_error = [0.12, 0.13, 0.2, 0.1]
# 5.6: define width, and height as figsize=(4,5) DPI and adjust plot dpi=100
# 5.7: Give a font size of 14

import matplotlib.pyplot as plt

# Data points
X = [1, 2, 3, 4]
y = [20, 21, 20.5, 20.8]
y_error = [0.12, 0.13, 0.2, 0.1]

# Create the figure with width=4, height=5, and dpi=100
plt.figure(figsize=(4, 5), dpi=100)

# Plot with error bars
plt.errorbar(
    X, y,
    yerr=y_error,
    fmt='o-',             
    color='blue',
    ecolor='red',
    elinewidth=2,
    capsize=5
)

# Add labels and title with font size 14
plt.xlabel("X-axis", fontsize=14)
plt.ylabel("Y-axis (Values)", fontsize=14)
plt.title("Simple Plot with Error Bars", fontsize=14)

# Show the plot
plt.show()
```

```python
# 5.8: Draw a scatter graph of any 50 random values of the x and y axis

import numpy as np
import matplotlib.pyplot as plt

# Generate 50 random values for x and y
x = np.random.rand(50) * 100  
y = np.random.rand(50) * 100

# Create scatter plot
plt.figure()
plt.scatter(x, y, color="blue", marker="o", s=60)

# Add labels and title
plt.xlabel("X-axis (Random Values)")
plt.ylabel("Y-axis (Random Values)")
plt.title("Scatter Plot of 50 Random Points")

# Show the plot
plt.show()
```

#### 5.9: Create a dataframe from the following data
'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
'female': [0, 1, 1, 0, 1],
'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, 2, 3],
'postTestScore': [25, 94, 57, 62, 70]

```python
import pandas as pd

# Define the data as a dictionary
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'female': [0, 1, 1, 0, 1],
    'age': [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

# Create the DataFrame
df = pd.DataFrame(data)
df
```

```python
# Draw a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age

import matplotlib.pyplot as plt

plt.figure()

# Use age to scale marker size (multiply by factor for visibility)
plt.scatter(df['preTestScore'], df['postTestScore'],    
            c='blue',        
            alpha=0.6,       
            edgecolor='black')

plt.xlabel("Pre-Test Score")
plt.ylabel("Post-Test Score")
plt.title("Scatterplot of Pre vs Post Test Scores (Sized by Age)")

plt.show()
```

```python
# 5.10: Draw a Scatterplot from the data in question 9 of preTestScore and postTestScore with the size = 300 and the color determined by sex

#Define colors based on sex (female=1 â†’ red, male=0 â†’ blue)
colors = df['female'].map({0: 'blue', 1: 'red'})

plt.figure()
plt.scatter(df['preTestScore'], df['postTestScore'],
            s=300,          # fixed size
            c=colors)      # color by sex

plt.xlabel("Pre-Test Score")
plt.ylabel("Post-Test Score")
plt.title("Scatterplot of Pre vs Post Test Scores (Color by Sex)")

# Add legend
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Male', markerfacecolor='blue', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Female', markerfacecolor='red', markersize=10)
]
plt.legend(handles=legend_elements, title="Sex")

plt.show()
```
