```python
import hashlib
import re

def encrypt_reference_id(ref_id):
    """Encrypt Reference ID using SHA-256 hashing."""
    return hashlib.sha256(ref_id.encode()).hexdigest()

def validate_reference_id(ref_id):
    """
    Validate Reference ID:
    - Must be 12 characters long
    - Can contain letters, digits, and optionally special characters
    """
    pattern = r'^[A-Za-z0-9@#$%^&+=]{12}$'  # allows alphanumeric + some special chars
    return re.match(pattern, ref_id) is not None

def main():
    # Step 1: Read input from user
    ref_id = input("Enter your 12-character Reference ID: ")

    # Step 2: Validate Reference ID
    if not validate_reference_id(ref_id):
        print("‚ùå Invalid Reference ID. It must be 12 characters (letters, digits, or allowed special chars).")
        return

    # Step 3: Encrypt Reference ID
    encrypted_id = encrypt_reference_id(ref_id)
    print(f"‚úÖ Encrypted Reference ID: {encrypted_id}")

    # Step 4 (Enhancement): Option for decryption (Note: SHA-256 is one-way, cannot decrypt)
    choice = input("Do you want to verify (simulate decryption)? (yes/no): ").lower()
    if choice == "yes":
        verify_id = input("Re-enter your Reference ID for verification: ")
        if encrypt_reference_id(verify_id) == encrypted_id:
            print("üîí Verification successful! Reference ID matches.")
        else:
            print("‚ùå Verification failed! Reference ID does not match.")

if __name__ == "__main__":
    main()
```
