#### 3. A website requires a user to input a username and password to register. Write a program to check the validity of the password given by the user. Following are the criteria for checking password:

1. At least 1 letter between [a-z]
2. At least 1 number between [0-9]
3. At least 1 letter between [A-Z]
4. At least 1 character from [$#@]
5. Minimum length of transaction password: 6
6. Maximum length of transaction password: 12
Hint: In the case of input data being supplied to the question, it should be assumed to be a console input

```python
import re

def is_valid_password(password):
    # Check length
    if len(password) < 6 or len(password) > 12:
        return False
    
    # Check at least one lowercase letter
    if not re.search("[a-z]", password):
        return False
    
    # Check at least one uppercase letter
    if not re.search("[A-Z]", password):
        return False
    
    # Check at least one digit
    if not re.search("[0-9]", password):
        return False
    
    # Check at least one special character from [$#@]
    if not re.search("[$#@]", password):
        return False
    
    return True


# Console input
username = input("Enter your username: ")
password = input("Enter your password: ")

if is_valid_password(password):
    print("Registration successful")
else:
    print("Invalid password. Please follow the password rules.")
```

#### 7.Please write a program thatcountsand printsthe numbers of each character in a string input by the console.Example: If the following string is given as input to the program:abcdefgabcThen, the output of the program should be:a,2c,2b,2e,1d,1g,1f,1

```python
# Accept string input from the console
s = input("Enter a string: ")

# Dictionary to store character counts
char_count = {}

# Count each character
for ch in s:
    char_count[ch] = char_count.get(ch, 0) + 1

# Print results in the required format
output = ""
for ch, count in char_count.items():
    output += f"{ch},{count}"

print(output)
```

#### 9.By using list comprehension, please write a program to print the list after removing the value 24 in [12,24,35,24,88,120,155].

```python
# Original list
a = [12, 24, 35, 24, 88, 120, 155]

# Remove all 24s using list comprehension
result = [x for x in a if x != 24]

print("List after removing 24:", result)
```

#### 10. By using list comprehension, please write a program to print the list after removing the 0th,4th,and 5th numbers in [12,24,35,70,88,120,155].

```python
# Original list
a = [12, 24, 35, 70, 88, 120, 155]

# Remove elements at positions 0, 4, and 5 using list comprehension
result = [a[i] for i in range(len(a)) if i not in (0, 4, 5)]

print("List after removing 0th, 4th, and 5th elements:", result)
```

#### 12. Write  a  program  to  compute  1/2+2/3+3/4+...+n/n+1  with  a  given  n  input  by console (n>0).Example:If the following n is given as input to the program:5Then, the output of the program should be:3.55

```python
# Program to compute 1/2 + 2/3 + 3/4 + ... + n/(n+1)

# Accept input from console
n = int(input("Enter a positive integer n: "))

# Ensure n > 0
if n <= 0:
    print("n must be greater than 0")
else:
    total = 0
    for i in range(1, n + 1):
        total += i / (i + 1)
    
    print("Result:", round(total, 2))
```
