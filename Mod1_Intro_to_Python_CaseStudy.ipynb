# 1. Write a program that will find factors of the given number and find whether the factor is even or odd </br>
Hint: Use Loop with if-else statements

```python
def find_factors_and_check_parity(n):
    factors = []
    for i in range(1, n + 1):
        if n % i == 0:
            parity = "Even" if i % 2 == 0 else "Odd"
            factors.append((i, parity))
    return factors


number = int(input("Enter a number: "))
result = find_factors_and_check_parity(number)

print(f"Factors of {number} and their parity:")
for factor, parity in result:
    print(f"{factor} -> {parity}")
```

# 2. Write a code that accepts a sequence of words as input and prints the words in a sequence after sorting them alphabetically

```python
def sort_words():
    # Accept input from the user
    words = input("Enter a sequence of words separated by spaces: ")
    
    # Split the input into a list of words
    word_list = words.split()
    
    # Sort the list alphabetically
    word_list.sort()
    
    # Print the sorted words
    print("Sorted words:")
    for word in word_list:
        print(word)

# Run the function
sort_words()
```

# 3. Write a program, which will find all the numbers between 1000 and 3000 (both included) such that each digit of a number is an even number. The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
def find_even_digit_numbers():
    result = []
    for num in range(1000, 3001):  # Include 3000
        num_str = str(num)
        if all(int(digit) % 2 == 0 for digit in num_str):  # Check if all digits are even
            result.append(num_str)
    
    # Print as comma-separated sequence
    print(",".join(result))

# Run the function
find_even_digit_numbers()
```

# 4- Write a program that accepts a sentence and calculates the number of letters and digits.

```python

def count_letters_and_digits():
    # Accept input from the user
    sentence = input("Enter a sentence: ")
    
    # Initialize counters
    letters = 0
    digits = 0
    
    # Iterate through each character
    for char in sentence:
        if char.isalpha():
            letters += 1
        elif char.isdigit():
            digits += 1
    
    # Display the results
    print(f"Number of letters: {letters}")
    print(f"Number of digits: {digits}")

# Run the function
count_letters_and_digits()
```

# 5- Design a code that will find whether the given number is a Palindrome number or not

```python
def is_palindrome_number():
    # Accept input from the user
    num = input("Enter a number: ")
    
    # Check if the number is the same when reversed
    if num == num[::-1]:
        print(f"{num} is a Palindrome number.")
    else:
        print(f"{num} is NOT a Palindrome number.")

# Run the function
is_palindrome_number()
```
