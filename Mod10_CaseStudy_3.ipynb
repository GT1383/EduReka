4. Compute the mean squared error by comparing the actual class and predicted high-level class.

```python
# Load the file “zoo.data” and look at the info and the first five rows. 
# The first column denotes the animal name and the last one specifies a high-level class for the corresponding animal.

import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import mean_squared_error

# Load dataset
df = pd.read_csv("zoo.csv")

# Look at info and first five rows
print(df.info())
df.head()

```

```python
# Find out the unique number of high-level cclasses

unique_classes = df['class_type'].nunique()
print("Unique number of high-level classes:", unique_classes)
```

```python
# Use the 16-intermediate feature and perform an agglomerative clustering. 
# [ Hint: Refer to the agglomerative clustering module in scikit learn and set the number of clusters appropriately]


# Use the 16 intermediate features (exclude animal_name and class)
X = df.iloc[:, 1:-1]   
y = df['class_type']   

# Agglomerative Clustering
agg_cluster = AgglomerativeClustering(n_clusters=unique_classes, linkage='ward')
y_pred = agg_cluster.fit_predict(X)
```

```python
# Compute the mean squared error by comparing the actual class and predicted high-level class.

mse = mean_squared_error(y, y_pred)
print("Mean Squared Error between actual and predicted classes:", mse)
```
