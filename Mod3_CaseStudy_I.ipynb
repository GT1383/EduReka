#### 1. A Robot moves in a Plane starting from the origin point (0,0). The robot can move UP, DOWN, LEFT, or RIGHT. The trace of Robot movement is as given following: UP 5 DOWN 3 LEFT 3 RIGHT 2 The numbers after directions are steps. Write a program to compute the distance current position after a sequence of movements. Hint: Use the math module.

```python
import math

def robot_distance(movements):
    # Starting point
    x, y = 0, 0
    
    # Process each movement
    for direction, steps in movements:
        if direction.upper() == "UP":
            y += steps
        elif direction.upper() == "DOWN":
            y -= steps
        elif direction.upper() == "LEFT":
            x -= steps
        elif direction.upper() == "RIGHT":
            x += steps
    
    # Compute Euclidean distance from origin
    distance = math.sqrt(x**2 + y**2)
    return (x, y, distance)


# Example trace of movements
movements = [("UP", 5), ("DOWN", 3), ("LEFT", 3), ("RIGHT", 2)]

final_x, final_y, dist = robot_distance(movements)

print(f"Final Position: ({final_x}, {final_y})")
print(f"Distance from Origin: {dist:.2f}")
```

#### 2. Data of XYZ company is stored in the sorted list. Write a program for searching specific data from that list.Hint: Use if/elif to deal with conditions.


```python
def binary_search(sorted_list, target):
    low = 0
    high = len(sorted_list) - 1
    
    while low <= high:
        mid = (low + high) // 2
        
        if sorted_list[mid] == target:
            return mid   # found at index mid
        elif sorted_list[mid] < target:
            low = mid + 1   # search right half
        else:
            high = mid - 1  # search left half
    
    return -1   # not found


# Example usage
data = ["Amit", "Bob", "Pankaj", "David", "Gaurav", "Rich"]  # sorted list
target = input("Enter the name to search: ")

index = binary_search(data, target)

if index != -1:
    print(f"‚úÖ {target} found at position {index}.")
else:
    print(f"‚ùå {target} not found in the list.")
```

#### 4. Write a program to find the distance between two locations when their latitude and longitudes are given. Hint: Use the math module.

```python
import math

def haversine(lat1, lon1, lat2, lon2):
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    # Radius of Earth in kilometers
    R = 6371.0
    distance = R * c
    return distance

# Example usage
lat1 = float(input("Enter latitude of location 1: "))
lon1 = float(input("Enter longitude of location 1: "))
lat2 = float(input("Enter latitude of location 2: "))
lon2 = float(input("Enter longitude of location 2: "))

dist = haversine(lat1, lon1, lat2, lon2)
print(f"Distance between locations: {dist:.2f} km")
```

#### 5. Design software for bank systems. There should be options like cash withdrawal, cash credit, and a change password. According to user input, the software should provide the required output. Hint: Use if else statements and functions.

```python
def cash_withdraw(balance):
    amount = float(input("Enter amount to withdraw: "))
    if amount <= balance:
        balance -= amount
        print(f"‚úÖ Withdrawal successful. Remaining balance: {balance}")
    else:
        print("‚ùå Insufficient balance.")
    return balance

def cash_credit(balance):
    amount = float(input("Enter amount to credit: "))
    balance += amount
    print(f"‚úÖ Amount credited successfully. New balance: {balance}")
    return balance

def change_password(password):
    old_pass = input("Enter your old password: ")
    if old_pass == password:
        new_pass = input("Enter your new password: ")
        confirm_pass = input("Confirm your new password: ")
        if new_pass == confirm_pass:
            password = new_pass
            print("‚úÖ Password changed successfully.")
        else:
            print("‚ùå Passwords do not match.")
    else:
        print("‚ùå Incorrect old password.")
    return password

def main():
    balance = 1000.0   # initial balance
    password = "bank123"  # default password
    
    while True:
        print("\n--- Bank System Menu ---")
        print("1. Cash Withdrawal")
        print("2. Cash Credit")
        print("3. Change Password")
        print("4. Exit")
        
        choice = input("Enter your choice (1-4): ")
        
        if choice == "1":
            balance = cash_withdraw(balance)
        elif choice == "2":
            balance = cash_credit(balance)
        elif choice == "3":
            password = change_password(password)
        elif choice == "4":
            print("üëã Thank you for using our bank system.")
            break
        else:
            print("‚ùå Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

#### 6. Write a program that will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
# Program to find numbers divisible by 7 but not multiples of 5
# between 2000 and 3200 (inclusive)

result = []
for num in range(2000, 3201):  # 3201 because range upper bound is exclusive
    if num % 7 == 0 and num % 5 != 0:
        result.append(str(num))

# Print the numbers as a comma-separated sequence
print(",".join(result))
```

#### 7. Write a program that can compute the factorial of a given numb. Use recursion to find it. Hint: Suppose the following input is supplied to the program: 8 Then, the output should be: 40320

```python
# Recursive function to compute factorial
def factorial(n):
    if n == 0 or n == 1:   # Base case
        return 1
    else:
        return n * factorial(n - 1)  # Recursive case

# Take input from user
num = int(input("Enter a number: "))

# Compute and print factorial
print(f"Factorial of {num} is: {factorial(num)}")
```

#### 8. Write a program that calculates and prints the value according to the given formula:
#### Q = Square root of [(2 * C * D)/H]
#### Following are the fixed values of C and H: C is 50. H is 30.
#### D is the variable whose values should be input to your program in a comma separated sequence.
 #### Example:
#### Let us assume the following comma-separated input sequence is given to the
#### program:
#### 100,150,180
#### The output of the program should be:
#### 18,22,24

```python
import math

# Fixed values
C = 50
H = 30

# Take input from user
D_values = input("Enter comma-separated values of D: ")

# Split input into list of integers
D_list = [int(x) for x in D_values.split(",")]

# Compute Q for each D
result = []
for D in D_list:
    Q = int(round(math.sqrt((2 * C * D) / H)))
    result.append(str(Q))

# Print results as comma-separated sequence
print(",".join(result))
```

#### 11. Write a program that accepts sequence of lines as input and prints the lines after making all characters in the sentence capitalized.
#### Suppose the following input is supplied to the program:
#### Hello world
#### Practice makes perfect
#### Then, the output should be:  HELLO WORLD PRACTICE MAKES PERFECT

```python
# Program to capitalize all characters in input lines

# Collect lines from user until they press Enter without typing anything
lines = []
while True:
    line = input("Enter a line (press Enter to stop): ")
    if line:   # if line is not empty
        lines.append(line.upper())  # convert to uppercase
    else:
        break

# Print the result
for l in lines:
    print(l)
```

#### 12. Write a program that accepts a sequence of whitespace-separated words as input and prints the words after removing all duplicate words and sorting them alphanumerically. Suppose the following input is supplied to the program:  hello world and practice makes perfect and hello world again
#### Then, the output should be: again and hello makes perfect practice world.


```python
# Program to remove duplicates and sort words alphanumerically

# Take input from user
words = input("Enter whitespace-separated words: ")

# Split into list of words
word_list = words.split()

# Remove duplicates using set, then sort
unique_sorted_words = sorted(set(word_list))

# Print the result
print(" ".join(unique_sorted_words))
```

#### 13.Write a program that accepts a sequence of comma separated 4 digit binary numbers as its input and then check whether they are divisible by 5 or not. The numbers that are divisible by 5 are to be printed in a comma-separated sequence.
#### Example: 0100,0011,1010,1001
#### Then the output should be: 1010

```python
# Program to check binary numbers divisible by 5

# Take input from user
binary_numbers = input("Enter comma-separated 4-digit binary numbers: ")

# Split input into list
binary_list = binary_numbers.split(",")

# Check divisibility by 5
result = []
for b in binary_list:
    decimal_value = int(b, 2)  # Convert binary to decimal
    if decimal_value % 5 == 0:
        result.append(b)

# Print result
print(",".join(result))
```

#### 14.Write a program that accepts a sentence and calculates the number of upper case letters and lower case letters.
#### Suppose the following input is supplied to the program:  Hello world!
#### Then, the output should be:  UPPER CASE 1  LOWER CASE 9


```python
# Program to count uppercase and lowercase letters in a sentence

# Take input from user
sentence = input("Enter a sentence: ")

# Initialize counters
upper_count = 0
lower_count = 0

# Loop through each character
for char in sentence:
    if char.isupper():
        upper_count += 1
    elif char.islower():
        lower_count += 1

# Print results
print("UPPER CASE", upper_count)
print("LOWER CASE", lower_count)
```

#### 15. Give an example of the fsum and sum function of the math library

```python
import math

# A list of floating-point numbers
numbers = [0.1] * 10  # ten times 0.1

# Using built-in sum
normal_sum = sum(numbers)

# Using math.fsum
accurate_sum = math.fsum(numbers)

print("Using sum():", normal_sum)
print("Using math.fsum():", accurate_sum)
```
